 Dictinaries
* Items in dictionaries can only string, int, tuple type and not dictionary, set or list

```python
d = dict()
d[{1, 2}] = 3  # ERROR
d[{1: 2}] = 3  # ERROR
d[[1, 2]] = 3  # ERROR
```

Booleans
* Bools return false when the list is empty, number is zero, string is empty etc.

```python
state1 = bool(0) or bool([]) or bool({}) or bool("")  # state is false
state3 = bool({0}) and bool([0])  # state is True
print(state1)
print(state3)
```

Map

We can also go over two iterables


```python
items = [1, 2, 3, 4, 5]
squared1 = list(map(lambda x: x ** 2, items))
squared2 = []
for i in items:
    squared2.append(i ** 2)
print(squared1 == squared2)
```

Reduce

reduce can receive a third argument, that will act as an initializer and be placed before the other items


```python
from functools import reduce

print(reduce((lambda x, y: x * y), [1, 2, 3, 4]))  # 1*2=2, 2*3= 6, 6*4 = 24
print(reduce(lambda x, y: x + "*" + y, ["is", "fun"], "intro"))  # OUTPUT "intro*is*fun"
```
